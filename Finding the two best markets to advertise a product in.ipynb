## FINDING BEST MARKETS TO ADVERTISE AN E-LEARNING PRODUCT IN


Let's say I have an e-learning product called ABC that has courses mainly focused on Web Development, Mobile Development, and Data Science. There are also other courses in it like Mobile development, UX designing, etc., All the courses are entirely in English and each course costs $60 per month. Through data analysis, I'll try to find the best markets to advertise my product in.

For this analysis, I'll be using [FreeCodeCamp's](https://www.freecodecamp.org/) New coder Survey dataset. The dataset is available in [Kaggle](https://www.kaggle.com/freecodecamp/freecodecamp-2018-new-coder-survey-of-30k-devs?). 

FreeCodeCamp is a free e-learning platform that offers courses on a variety of domains. Their survey attracted many new coders with varying interests, which is ideal for  my analysis


### READING IN THE DATASET

```python
#Importing necessary modules
import pandas as pd

#Reading in the dataset
fcc = pd.read_csv('fcc.csv')
print(fcc.shape)
pd.options.display.max_columns = 140
fcc.head()
```

 ### CHANGING THE COLUMN NAMES

```python
import re

#Crating a list variable to store the names of the columns starting with 'job_intr_'
columns = []
for col in fcc.columns:
    if re.search('^job_intr_', col):
        columns.append(col)
#Printing the original column names
        print(columns)


```

```python
#Renaming the columns
fcc.rename(columns = {'job_intr_fllstck' : 'Full-Stack Web Developer',
                      'job_intr_backend' : 'Back-End Web Developer',
                      'job_intr_frntend' : 'Front-End Web Developer',
                      'job_intr_mobile' : 'Mobile Developer',
                      'job_intr_devops' : 'DevOps',
                      'job_intr_datasci': 'Data Scientist',
                      'job_intr_teacher' : 'Teacher/Professor',
                      'job_intr_qa_engn' : 'Quality Assurance Engineer',
                      'job_intr_ux_engn' : 'User Experience Designer',
                      'job_intr_projm' : 'Project Manager',
                      'job_intr_gamedev' : 'Game Developer',
                      'job_intr_infosec' : 'Information Security',
                      'job_intr_dataengn' : 'Data Engineer',
                      'job_intr_other' : 'Other'
                     },inplace=True)
columns = ['Full-Stack Web Developer', 'Back-End Web Developer',
       'Front-End Web Developer', 'Mobile Developer', 
           'DevOps','Data Scientist','Teacher/Professor',
           'Quality Assurance Engineer','User Experience Designer',
           'Project Manager','Game Developer',
           'Information Security','Data Engineer','Other']
```

### CREATING JOB ROLE INTEREST COLUMN

I am going to create a new column to store the names of the roles each survey participant is interested in, by aggregating the job interest columns.

```python
#importing numpy
import numpy as np
fcc['job_role_interest'] = ''
for index, row in fcc.iterrows():
    for item in columns:
        if row[item]==1.0:
            if fcc.at[index,'job_role_interest'] == '':
                fcc.at[index,'job_role_interest'] += item
            else:
                fcc.at[index,'job_role_interest'] += ", " + item
                    
```

```python
#Replacing empty srings with NaN in 'job_role_interest' column
fcc['job_role_interest'] = fcc['job_role_interest'].replace(r'^\s*$', np.NaN, regex=True)
```

```python
fcc.head()
```

### CHECKING FOR SAMPLE REPRESENTATIVITY

As mentioned earlier, most of the courses on ABC are based on Web development and data science. It is important to clarify whether the dataset has the right categories of people for our purposes. In the next few sections, I will be analyzing the population of new coders who are interested in the subjects that are available on ABC.

If a person is interested in a particular domain, there is a high chance that the person takes a course on that subject. So I'll start with analyzing the 'job_role_interest_column'. This column describes the role each participant is interested in working in.


```python
#Frequency distribution
fcc['job_role_interest'].value_counts(normalize=True)*100
```

```python
print('Number of participants interested in Web Development',fcc['job_role_interest'].str.contains('Web Developer').sum())
print('Number of participants interested in Data Science',fcc['job_role_interest'].str.contains('Data Scientist').sum())
print('Number of participants interested in Mobile Development',fcc['job_role_interest'].str.contains('Mobile Developer').sum())
```

From the frequency table, we can see that
- A lot of people are interested in full-stack and front-end development (web development).
- A few people are interested in data science and mobile development.
- A few people are interested in domains other than web development, data science, and mobile development.

It is interesting to see that many people are interested in more that one domain. In the next code block, I'll be splitting the strings in each row of the 'job_role_interest' column to find the number of options for each participant. And then, I'll generate a frequency table describing the number of options.


```python
#Dropping null values
role_interest_no_null = fcc['job_role_interest'].dropna()
#Splitting each string in the 'job_role_interest_column'
interest = role_interest_no_null.str.split(', ')
#Generating a frequenc table
num_of_options = interest.apply(lambda x : len(x))
num_of_options.value_counts(normalize=True).sort_index()*100
```

It turns out that only 34.2% of the participants have a clear idea about what programming niche they'd like to work in. The vast majority of the respondents have mixed interests. The fact that the new coders have mixed interests might be good as ABC offers courses on various subjects. Since most of the ABC's courses are on web development and data science, I'll try to find how many new coders are interested in at least one of the two courses.

```python
web_mob_or_data = role_interest_no_null.str.contains('Web Developer|Data Scientist|Mobile Developer')
frequency_table = web_mob_or_data.value_counts(normalize=True)*100
print(frequency_table)
```

```python
#Visualizing the frequenc table
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
frequency_table.plot.bar()
plt.title('Most participants are interested in Web Development\n or Data Science or Mobile Development',y=1.1)
plt.xticks([0,1],['Web Development\nor Data Science\nor Mobile Development','Other'],rotation=0)
plt.ylabel('Percentage')
plt.ylim([0,100])
plt.show()
```

From the bar plot, we can see that 88.6% of the participants are interested in Web Development or Data Science, or Mobile Development. This indicates that the sample is strongly representative of my population of interest since the goal is to advertise to people interested in all kinds of programming niches but mostly web development, data science, and mobile development.

### LOCATION AND DENSITIES

I'll now find the number of participants in each country. To make sure the sample remains representative, I'll remove the participants who've not answered what role they're interested in.

```python
#Isolating the participants
fcc_rep = fcc[fcc['job_role_interest'].notnull()].copy()

#Frequency table
absolute_frequencies = fcc_rep['country_live'].value_counts()
relative_frequencies = fcc_rep['country_live'].value_counts(normalize=True)*100

#Displaying frequency tables
pd.DataFrame(data = {'Absolute Frequency' : absolute_frequencies, 'Percentage' : relative_frequencies})
```

We can see that 37.8% of potential customers belong to the USA and 14.4% belong to India, which is not too far from the United Kingdom (4.05%).

### SPENDING MONEY FOR LEARNING

The money_for_learning column describes the amount of money (in dollars) spent by the participants from the moment they started coding until the moment they completed the survey. I'll assume that ABC sells subscriptions for a price of $50 per month. And for this reason, I'm interested in finding out how much money people are willing to spend per month on learning.

I'll narrow down my analysis to four countries: The USA, India, United Kingdom, and Canada for these two reasons:
1. These countries have a high frequency in the frequency table.
2. ABC's courses are entirely in English and English is an official language in all these countries. The more people know English, the better the chances to target the right people with our ads.

Now I'll find the money spent per month by each participant by dividing the 'money_for_learning' column by the 'months_programming' column. The only problem is that some participants have answered '0' for 'months_programming'. So I’ll replace '0' with '1' to avoid dividing by '0'.



```python
#Replacing '0' with '1' in 'months_programming' column
fcc_rep['months_programming'].replace(0,1,inplace=True)

#Money sent per month
fcc_rep['money_per_month'] = fcc_rep['money_for_learning']/fcc_rep['months_programming']
```

```python
#Removing the rows with null values
fcc_rep['money_per_month'].isnull().sum()
```

There are 79 null values in 'money_per_month'. I'll keep only the rows that do not have null values in the 'money_per_month' column.

```python
fcc_rep = fcc_rep[fcc_rep['money_per_month'].notnull()]
```

I am going to group the data by country to find the average amount of money that participants spend per month in each country.

```python
#Removing the rows with null values in 'country_live'
fcc_rep = fcc_rep[fcc_rep['country_live'].notnull()]

countries_group = fcc_rep.groupby('country_live').mean()
countries_group['money_per_month'][['United States of America', 'India', 'United Kingdom', 'Canada']]
```

The results for the United Kingdom and Canada are surprising when compared to the value we see for India. If we considered socio-economical metrics, we'd expect that people in the UK and Canada spend more on learning than people in India.

This might be correct or incorrect. It might be that we do not have enough representative data for the UK and Canada, or it might be due to outliers which makes the mean too large for India or too low for the UK and Canada. Next, I'll be analyzing the outliers to find answers to these questions.


### DEALING WITH EXTREME OUTLIERS

```python
#Isolating the 4 countries
country_4 = fcc_rep[fcc_rep['country_live'].str.contains('United States of America|^India|United Kingdom|Canada')]

#Box plots to visualize outliers
import seaborn as sns
sns.boxplot(x = 'country_live', y = 'money_per_month', data = country_4)
plt.title('Money spent per Month in each Country',y=1.1)
plt.xlabel('Country')
plt.ylabel('Money spent per Month (USD)')
plt.xticks(range(4),['India','USA','Canada','UK'])
plt.show()
```

There are some outliers in the data for India and the USA. The outliers in the USA are extreme. It indicates that two persons spend more than \\$50000 each month for learning. This seems extremely unlikely. So I'll remove the values that are greater than \\$20000 per month.

```python
#Isolating the participants who spend less than $20000 per month
fcc_rep = fcc_rep[fcc_rep['money_per_month']<20000]
```

Now I'll compute the mean values and box plot again to see if there are any outliers.

```python
countries_group = fcc_rep.groupby('country_live').mean()
countries_group['money_per_month'][['United States of America', 'India', 'United Kingdom', 'Canada']]
```

```python
country_4 = fcc_rep[fcc_rep['country_live'].str.contains('United States of America|^India|United Kingdom|Canada')]

#Box plots to visualize outliers
import seaborn as sns
sns.boxplot(x = 'country_live', y = 'money_per_month', data = country_4)
plt.title('Money spent per Month in each Country',y=1.1)
plt.xlabel('Country')
plt.ylabel('Money spent per Month (USD)')
plt.xticks(range(4),['India','USA','Canada','UK'])
plt.show()
```

There are a few extreme outliers for all the 4 countries. But it is still unclear what causes these outliers. Maybe these participants attended several boot camps in a month which tend to be very expensive. 
Now I'll move on to examining the outliers.


### OUTLIERS FOR INDIA

```python
#Outliers for India
india_outliers = country_4[(country_4['country_live'] == 'India') & (country_4['money_per_month']>=5000)]
india_outliers
```

It seems that none of them have attended a boot camp. It is hard to figure out from the data whether they spent that much money on learning. The question asked on the survey was "Aside from university tuition, about how much money have you spent on learning to code so far (in US dollars)?". So they might have misunderstood and thought university tuition fees are included. So it is safe to remove these rows.

NOTE: You can find the raw dataset [here](https://raw.githubusercontent.com/freeCodeCamp/2018-new-coder-survey/master/raw-data/2018-new-coder-survey.csv). This dataset contains the uncleaned column names, where you can find the question.


```python
#Removing the outliers for India
country_4 = country_4.drop(india_outliers.index)
```

### OUTLIERS FOR USA

```python
#Outliers for USA
us_outliers = country_4[(country_4['country_live'] == 'United States of America') & (country_4['money_per_month'] >= 7000)]
us_outliers
```

```python
#Checking how many people have attended the bootcamp
print('Number of people who have attended bootcamp = ',us_outliers[us_outliers['bootcamp_attend'] == 1.0].shape[0])
print('Number of people who have not attended bootcamp = ',us_outliers[us_outliers['bootcamp_attend'] == 0.0].shape[0])
```

Out of the 31 extreme outliers, 16 people have attended the boot camp, which might be the reason for high 'money_per_month'. Whereas for the other 15, it's hard to find the reason. So it's safe to remove these rows where participants spend more than //$7000 and have not attended a Bootcamp.

Also, most of these outliers have 'months_programming' values <3. This shows that the participants might have paid huge amounts of money, in the beginning, to enroll in boot camps they planned to attend for several months. But at the time of the survey, they would have only completed a few months of Bootcamp. This explains the high 'money_per_month' values because of lower 'months_programming' value. They probably didn't spend any money on learning after the survey. So I'll also remove those outliers too.

I'll remove the rows where:
1. Participants who didn't attend boot camps
2. Participants who had been programming for three months or less at the time of the survey.


```python
#Removing the participants who've not attended a bootcamp
no_bootcamp = country_4[(country_4['country_live'] == 'United States of America') & (country_4['money_per_month'] >=7000) & (country_4['bootcamp_attend'] == 0.0)]
country_4 = country_4.drop(no_bootcamp.index)

#Removing the participants who've been programming for less than three months
less_than_3_months = country_4[(country_4['country_live'] == 'United States of America') & (country_4['money_per_month'] >= 7000) & (country_4['months_programming']<3)]
country_4 = country_4.drop(less_than_3_months.index)
```

### OUTLIERS FOR CANADA

```python
#Outliers for Canada
canada_outliers = country_4[(country_4['country_live'] == 'Canada') & (country_4['money_per_month'] > 7500)]
canada_outliers
```

One participant has not attended a boot camp. So I'll remove this row. The other participant has attended a boot camp and has been programming for only 1 month when he completed the survey. This explains the high value for money_per_month. So it is safe to remove this row because we do not know if the participant spent any money on learning after the survey.

```python
#Removing the outliers for Canada
country_4 = country_4.drop(canada_outliers.index)
```

### OUTLIERS FOR UK

```python
#Outliers for UK
uk_outliers = country_4[(country_4['country_live'] == 'United Kingdom') & (country_4['money_per_month'] > 7500)]
uk_outliers
```

3 out of 4 participants have attended a boot camp. I'll remove the participant who has not attended a boot camp. The three participants who have attended a Bootcamp have been programming only for a month when they completed the survey. That explains it. So I'll remove these rows too.

```python
#Removing the outliers
country_4 = country_4.drop(uk_outliers.index)
```

Now, I'll recompute the mean values and generate the box plot.

```python
country_4.groupby('country_live').mean()['money_per_month']
```

```python
sns.boxplot(x = 'country_live', y = 'money_per_month', data = country_4)
plt.title('Average Money spent per month on learning in each country', y = 1.1)
plt.xlabel('Country')
plt.ylabel('Money Per Month (USD)')
```
